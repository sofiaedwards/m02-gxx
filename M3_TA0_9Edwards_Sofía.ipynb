##  游댃 **Ciclos** 游댃

Algo que puede hacer Python muy bien (a diferencia de nosotros) es hacer acciones repetitivas y fomes. Ellas se hacen con **ciclos**. Los ciclos son funciones que se ocupan en un formato parecido a los condicionales, donde todo el c칩digo dentro de la indentaci칩n ser치 **repetido hasta que se cumpla una condici칩n**. Hay **dos tipos de ciclos**:

### 游댁  **For i in range()** 游댁

En el primer tipo de ciclo, la condici쑕 de repetici쑕 es la cantidad de veces que se le pide dar vueltas (una vez, 3 veces, 1000). Este valor se declara dentro de `range()`.

```python
for i in range(3):        #se repetir치 3 veces
  print("hola!")

for j in range(5):        #se puede ocupar cualquier letra o palabra en for () in range()
  print("chao!")
```

La variable `i` o `j` en el c칩digo anterior es un contador de **cu치ntas vueltas ha dado.** Sin embargo, python cuenta **del 0 en adelante**. es decir, un ciclo que da 3 vueltas ir칤a del **0 al 2**. Por eso se dice que va del 0 al n-1 (el n칰mero m치ximo -1)

```python
for i in range(3):        #se repetir치 3 veces
  print("vuelta i n칰mero", i)

for j in range (5):
  print("vuelta j n칰mero", j)
```

Puedes colocar ciclos dentro de ciclos, lo que har칤a que el ciclo `j` se repita en cada vuelta `i`:

```python
for i in range(3):        #se repetir치 3 veces
  print("vuelta de i n칰mero", i)
  for j in range (5):     #se repetir 5 eces cada vuelta de i
    print("donde i es", i , "y j es", j )
```

Pero range() no solo puede tomar el n칰mero tope de vueltas, sino tambi칠n desde qu칠 n칰mero se parte. En vez de contar del 0 al 4, puede ser del 2 al 4. A diferencia del tope m치ximo, el tope m칤nimo s칤 se cuenta a s칤 mismo (se escribe (2,5), no (3,5)).

```python
for j in range (2,5):     #ir치 del 2 al 4
    print("j es", j )
```

Ahora aplicando variables se pueden realizar c칩digos muy interesantes. Un ejemplo cl치sico de c칩mo usar ciclos es un **contador**:

```python
num = int(input("Cuantos n칰meros contar칠?: "))

for i in range(1, num +1):    #cuenta del 1 al numero (si no lo haria del 0 al num-1)
  print("cuento", i)

print("wow, logr칠 contar hasta", num)
```

### 游뱕 M칩dulos % 游뱕

쯈u칠 pasa si quiero **contar de a n칰meros pares**? 쯢 de a tres?, 쯢 de a 22? Hay algunas maneras para que Python entienda que se est치 contando cada ciertos n칰meros y una de ellas es con su **m칩dulo** `%`.

Sup칩n que tienes una divisi칩n como **5/2**. en decimales eso es 2,5. Ya 5 que no es un m칰ltiplo de 2 te va a dar decimales. El m칩dulo cuenta los **n칰meros que le sobran al entero de la divisi칩n.**, en resumen 5/2  =  4/2 + 1/2  =  2 + 1/2 <- este sobrante es el m칩dulo

```python
print("el modulo de 5 en 2 es", 5%2, "porque le sobra", 5%2, "para ser una division entera")
print("a diferencia de 4 en 2 que es", 4%2, "el cual es 0 porque ya da un n칰mero entero")
```

En fin. El punto es que se puede contar cada ciertos n칰meros (cada cuatro por ejemplo) con i%4 == 0. Con ello, cada n칰mero multiplo de 4 ser치 considerado dentro de la condici칩n.

```python
num = int(input("Cuantos n칰meros contar칠?: "))

for i in range(1, num +1):    #cuenta del 1 al numero (si no lo haria del 0 al num-1)
  if i % 4 == 0:              #solo pasa si es multiplo de 4
    print("cuento el m칰ltiplo de cuatro", i)

print("wow, logr칠 contar hasta", num)
```

Y se puede guardar en una variable desde antes para ver cuantos n칰meros de ellos se contaron.

```python
num = int(input("Cuantos n칰meros contar칠?: "))

cuenta = 0

for i in range(1, num +1):    #cuenta del 1 al numero (si no lo haria del 0 al num-1)
  if i % 4 == 0:              #solo pasa si es multiplo de 4
    cuenta += 1               #cuando pasa el condicional le agrega 1 a la cantidad de numeros contados
    print("numero", i, "llevo", cuenta, "numeros contados")

print("wow, logr칠 contar hasta", num, "en total", cuenta, "numeros multiplos de 4")
```

### 游대 **While** 游대

La otra forma de realizar ciclos en con `while`. La condici칩n de ellas es seguir hasta el infinito a menos que una condici칩n cambie. Por ejemplo, le podemos decir que cuente hasta que el n칰mero sea mayor a 5.

```python
num = 0
while  num < 5:
  num += 1
  print("vamos en el n칰mero", num)

print("se lleg칩 hasta el n칰mero", num)
```

Se pueden realizar ciclos muy paracidos al anterior, como otro contador.

```python
num = int(input("Cuantos n칰meros contar칠?: "))

contador = 0
while contador < num:
  contador += 1
  print("vamos en el n칰mero", contador)

print("se lleg칩 hasta el n칰mero", contador)
```

Sin embargo, los `while` pueden ser **peligrosos**, porque si no colocas una condici칩n que los detenga, **seguir치n por la eternidad** y podr칤an generar una pantalla pegada en el computador 游땩.

Pero traen otros beneficios. Otras formas de ocupar `while` son con **condicionales.** Podemos colocar varias condiciones para que se detenga un ciclo.
