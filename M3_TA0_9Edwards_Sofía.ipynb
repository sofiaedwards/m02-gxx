##  🔄 **Ciclos** 🔄

Algo que puede hacer Python muy bien (a diferencia de nosotros) es hacer acciones repetitivas y fomes. Ellas se hacen con **ciclos**. Los ciclos son funciones que se ocupan en un formato parecido a los condicionales, donde todo el código dentro de la indentación será **repetido hasta que se cumpla una condición**. Hay **dos tipos de ciclos**:

### 🔂  **For i in range()** 🔂

En el primer tipo de ciclo, la condiciòn de repeticiòn es la cantidad de veces que se le pide dar vueltas (una vez, 3 veces, 1000). Este valor se declara dentro de `range()`.

```python
for i in range(3):        #se repetirá 3 veces
  print("hola!")

for j in range(5):        #se puede ocupar cualquier letra o palabra en for () in range()
  print("chao!")
```

La variable `i` o `j` en el código anterior es un contador de **cuántas vueltas ha dado.** Sin embargo, python cuenta **del 0 en adelante**. es decir, un ciclo que da 3 vueltas iría del **0 al 2**. Por eso se dice que va del 0 al n-1 (el número máximo -1)

```python
for i in range(3):        #se repetirá 3 veces
  print("vuelta i número", i)

for j in range (5):
  print("vuelta j número", j)
```

Puedes colocar ciclos dentro de ciclos, lo que haría que el ciclo `j` se repita en cada vuelta `i`:

```python
for i in range(3):        #se repetirá 3 veces
  print("vuelta de i número", i)
  for j in range (5):     #se repetirà 5 eces cada vuelta de i
    print("donde i es", i , "y j es", j )
```

Pero range() no solo puede tomar el número tope de vueltas, sino también desde qué número se parte. En vez de contar del 0 al 4, puede ser del 2 al 4. A diferencia del tope máximo, el tope mínimo sí se cuenta a sí mismo (se escribe (2,5), no (3,5)).

```python
for j in range (2,5):     #irá del 2 al 4
    print("j es", j )
```

Ahora aplicando variables se pueden realizar códigos muy interesantes. Un ejemplo clásico de cómo usar ciclos es un **contador**:

```python
num = int(input("Cuantos números contaré?: "))

for i in range(1, num +1):    #cuenta del 1 al numero (si no lo haria del 0 al num-1)
  print("cuento", i)

print("wow, logré contar hasta", num)
```

### 🤪 Módulos % 🤪

¿Qué pasa si quiero **contar de a números pares**? ¿o de a tres?, ¿o de a 22? Hay algunas maneras para que Python entienda que se está contando cada ciertos números y una de ellas es con su **módulo** `%`.

Supón que tienes una división como **5/2**. en decimales eso es 2,5. Ya 5 que no es un múltiplo de 2 te va a dar decimales. El módulo cuenta los **números que le sobran al entero de la división.**, en resumen 5/2  =  4/2 + 1/2  =  2 + 1/2 <- este sobrante es el módulo

```python
print("el modulo de 5 en 2 es", 5%2, "porque le sobra", 5%2, "para ser una division entera")
print("a diferencia de 4 en 2 que es", 4%2, "el cual es 0 porque ya da un número entero")
```

En fin. El punto es que se puede contar cada ciertos números (cada cuatro por ejemplo) con i%4 == 0. Con ello, cada número multiplo de 4 será considerado dentro de la condición.

```python
num = int(input("Cuantos números contaré?: "))

for i in range(1, num +1):    #cuenta del 1 al numero (si no lo haria del 0 al num-1)
  if i % 4 == 0:              #solo pasa si es multiplo de 4
    print("cuento el múltiplo de cuatro", i)

print("wow, logré contar hasta", num)
```

Y se puede guardar en una variable desde antes para ver cuantos números de ellos se contaron.

```python
num = int(input("Cuantos números contaré?: "))

cuenta = 0

for i in range(1, num +1):    #cuenta del 1 al numero (si no lo haria del 0 al num-1)
  if i % 4 == 0:              #solo pasa si es multiplo de 4
    cuenta += 1               #cuando pasa el condicional le agrega 1 a la cantidad de numeros contados
    print("numero", i, "llevo", cuenta, "numeros contados")

print("wow, logré contar hasta", num, "en total", cuenta, "numeros multiplos de 4")
```

### 🔁 **While** 🔁

La otra forma de realizar ciclos en con `while`. La condición de ellas es seguir hasta el infinito a menos que una condición cambie. Por ejemplo, le podemos decir que cuente hasta que el número sea mayor a 5.

```python
num = 0
while  num < 5:
  num += 1
  print("vamos en el número", num)

print("se llegó hasta el número", num)
```

Se pueden realizar ciclos muy paracidos al anterior, como otro contador.

```python
num = int(input("Cuantos números contaré?: "))

contador = 0
while contador < num:
  contador += 1
  print("vamos en el número", contador)

print("se llegó hasta el número", contador)
```

Sin embargo, los `while` pueden ser **peligrosos**, porque si no colocas una condición que los detenga, **seguirán por la eternidad** y podrían generar una pantalla pegada en el computador 😢.

Pero traen otros beneficios. Otras formas de ocupar `while` son con **condicionales.** Podemos colocar varias condiciones para que se detenga un ciclo.

```python
num = int(input("Cuantos números contaré?: "))

contador = 0
while contador < num and contador < 10:             #ahora va a continuar solo si es menor a num Y menor a 10
  contador += 1
  print("vamos en el número", contador)

print("se llegó hasta el número", contador)
```

Con ello puedes realizar **juegos por turnos**, donde terminan sólo si uno de los dos jugadores gana. Aquí un ejemplo de un juego de **ca-chi-pun** donde gana el mejor de 3:

```python

jugador_1 = input("nombre jugador 1: ")
jugador_2 = input("nombre jugador 2: ")

puntos_1 = 0
puntos_2 = 0

while puntos_1 < 3 and puntos_2 < 3:

  jugada_1 = input("jugador 1, escoje piedra/papel/tijeras: ")
  jugada_2 = input("jugador 2, escoje piedra/papel/tijeras: ")

  if jugada_1 == jugada_2:                      #en caso de que salga el mismo tipo
    print("empate!, se repite la jugada")

  elif jugada_1 == "piedra":                    #si el jugador 1 saca piedra
    if jugada_2 == "tijeras":                   # si el jugador 2 saca tijera gana el 1
      puntos_1 += 1
    elif jugada_2 == "papel":
      puntos_2 += 1
    else:
      print("algun input está malo, por favor volver a repetir")                             #si el jugador 2 saca papel gana el dos

  elif jugada_1 == "papel":                    #repetir con papel
    if jugada_2 == "piedra":
      puntos_1 += 1
    elif jugada_2 == "tijeras":
      puntos_2 += 1
    else:
      print("algun input está malo, por favor volver a repetir")

  elif jugada_1 == "tijeras":                     #repetir con tijeras
    if jugada_2 == "papel":
      puntos_1 += 1
    elif jugada_2 == "piedra":
      puntos_2 += 1
    else:
      print("algun input está malo, por favor volver a repetir")

  else:
      print("algun input está malo, por favor volver a repetir")
  print("puntaje actual", jugador_1, "tiene", puntos_1, "puntos,", jugador_2, "tiene", puntos_2, "puntos")

if puntos_1 ==  3:
  print(jugador_1, "ha ganado!")

elif puntos_2 ==  3:
  print(jugador_2, "ha ganado!")
```

*   Al final puede haber **un ganador o un empate**, en cuaolquier cado se debe **imprimir el resultado** con la cantidad de goles realizado por cada uno.




```python
contador = 0
puntos_alexis = 0
puntos_arturo = 0
while contador < 5:
  prob_alexis = random.randint(0,1)
  prob_arturo = random.randint(0,1)
  if prob_alexis > 0.5:
    puntos_alexis += 1
  if prob_arturo > 0.5:
    puntos_arturo += 1
  contador += 1

if puntos_arturo > puntos_alexis:
  print("Arturo Vidal ganó con",puntos_arturo, "puntos, y Alexis perdió con", puntos_alexis, "puntos")
elif puntos_arturo < puntos_alexis:
  print("Arturo Vidal ganó con",puntos_alexis, "puntos, y Arturo perdió con", puntos_arturo, "puntos")
else:
  print("empataron con", puntos_arturo, "puntos")

```
